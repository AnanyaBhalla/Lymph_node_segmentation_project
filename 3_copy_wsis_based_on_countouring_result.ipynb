```python
# Import modules
import os
import glob
import shutil
from PIL import Image
import cv2
import numpy as np
import openslide
```

Create a list with the names of the all slide names

```python
#Set the path to the folder containing all images
folder_path = "folder_with_all_images"

#Get the pathways and names of all the original WSIs
image_paths =glob.glob(folder_path,"/*/*")

#Filter so that only .ndpi images are selected
image_paths_ndpi = [i for i in image_paths if ".ndpi" in i]

#Initialize a list to store the names of WSIs
wsis_names = []

#Loop through the WSI image paths
for image_path in image_paths_ndpi:
    # Extract the base name of the image, Remove the last 5 characters(.ndpi)
    wsi_name = os.path.basename(image_path)[:-5].replace(" ", "_")  #Replace spaces with underscores
    # Append the processed slide name to the list
    wsis_names.append(wsi_name)

#print messege indicating that the process is complete
print("Done")
```

# Copy the incorrecly contoured images, labeled red. 
Tips: copy the red images before the green images. This allows you more time to annotate these images while the green copying code is running. 

```python
# Folder containing all the incorrectly contoured images
folder_red_mask = "folder/path/here" 

# Get all the pathway names for the incorrectly contoured WSIs
red_paths = glob.glob(folder_red_mask,"/*")

# Open a list for the names of the WSIs that were not correctly contoured
red_names = []

# Loop through the list of pathway names
for path in red_paths:
    #Extract the base name
    red_basename = os.path.basename(path)
    
    # Remove the "ending" at the end of the basename (adjust the number based on your requirement)
    mod_red_basename = red_basename[:-11] # Remove copied.ndpi
    
    # Append the modified basename to the list
    red_names.append(mod_red_basename)

#Print message to indicate that the process is complete
print("Complete")
```

```python
# Set the folder path for storing NDPI images that were not correctly annotated
red_ndpi_output_folder = "folder/path/to/red/wsis"

# Loop through pairs of WSI names and their corresponding image paths
for wsi_name, wsi_path in zip(wsis_names, image_paths_ndpi):
    
    # Loop through the names of incorrectly contoured WSIs
    for red_name in red_names:
        # Check if the current WSI path contains the red_name
        if red_name in wsi_path:
            # Customize the new file name
            new_file_name = f"{wsi_name}_copied.ndpi"
            
            # Create the new destination path
            new_destination_path = os.path.join(red_ndpi_output_folder, new_file_name)
            
            # Copy the file to the destination folder
            shutil.copy(wsi_path, new_destination_path)
            
            # Print a message indicating that the file has been copied
            print(f"Copied {wsi_path} to {new_destination_path}")
            
#Print message to indicate that the process is complete
print("Complete")
```

# Create copies of the correctly annotated lymph nodes, marked green. 
The incorrecly annotated images should alreayd be copied. This means that you can start to contoure these images while the correctly annotated images are now being copied. 

```python
# Folder containing all the correctly contoured images
folder_green_mask = "folder/path/here" 

# Get all the pathway names for the correctly contoured WSIs
green_paths = glob.glob(folder_gree_mask,"/*")

# Open a list for the names of the WSIs that were correctly contoured
green_names = []

# Loop through the list of pathway names
for path in green_paths:
    #Extract the base name
    green_basename = os.path.basename(path)
    
    # Remove the "ending" at the end of the basename (adjust the number based on your requirement)
    mod_green_basename = green_basename[:-11] # Remove copied.ndpi
    
    # Append the modified basename to the list
    green_names.append(mod_green_basename)

#Print message to indicate that the process is complete
print("Complete")
```

```python
# Set the folder path for storing NDPI images that were correctly annotated
green_ndpi_output_folder = "folder/path/here"

# Loop through pairs of WSI names and their corresponding image paths
for wsi_name, wsi_path in zip(wsis_names, image_paths_ndpi):
    
    # Loop through the names of correctly contoured WSIs
    for green_name in green_names:
        # Check if the current WSI path contains the green_name
        if green_name in wsi_path:
            # Customize the new file name
            new_file_name = f"{wsi_name}_copied.ndpi"
            
            # Create the new destination path
            new_destination_path = os.path.join(green_ndpi_output_folder, new_file_name)
            
            # Copy the file to the destination folder
            shutil.copy(wsi_path, new_destination_path)
            
            # Print a message indicating that the file has been copied
            print(f"Copied {wsi_path} to {new_destination_path}")
            
#Print message to indicate that the process is complete
print("Complete")
```
